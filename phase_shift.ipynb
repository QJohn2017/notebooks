Linear response
===============

The same treatment was applied to the linear response results, for the 3 cases that we currently have. Once again, the waves of the pump state are superimposed on the idler.

* $k_s = -0.4\, \mu m^{-1}$

```python
Image(filename="pngs/ks-0.4-2D.png")
```

Here is a slice though the $y=0$ line, showing the waves of the pump state (blue) and idler state (green).

```python
Image(filename="pngs/ks-0.4-1D.png")
```

The analysis of the phase shift is not straightforward here due to the fact that, close to the defect position, the idler state shows the superposition of the waves due to the pump ring and the ones due to the idler ring. *However*, since the group velocity of the idler is smaller than that of the pump (idler has a higher $k$-vector $k_i = 3.6\, \mu m^{-1}$), its associated waves will propagate a smaller distance away from the defect. Therefore we look at larger distances, and indeed see that the periodicity of the waves is equal to that of the pump, and in fact they *have opposite phase* (why?).

**Effect of simultaneous and opposite phase shift of signal and idler**

We have also looked at the effect of a simultaneous phase shift of $\frac{\pi}{4}$ ($-\frac{\pi}{4}$) of the signal (idler). This was done by multiplying the signal amplitude by $\frac{1}{\sqrt{2}}(1+i)$ and the idler by $\frac{1}{\sqrt{2}}(1-i)$. The resulting wave patterns are *unchanged*. For the 1 and 2-pump cases these phases cancel out in the linear response calculation. Perhaps for the OPO a similar thing occurs.
