Create our 'fortran code', which generates data every 5 seconds

```python
%%writefile fortran-code
#!/usr/bin/env python
import sys
import time
import numpy as np

n = int(sys.argv[1])

t = np.linspace(0,np.pi)

for omega in range(1,n):
    time.sleep(5)
    y = np.sin(omega * t)
    with open("data_%03i.dat" % omega, 'w') as f:
        np.save(f, y)
```

Define a function that plots a given data file

```python
def plot_fname(fname):
    with open(fname) as f:
        data = np.load(f)
    plt.plot(data)
    plt.title(fname)
    plt.show()
```

Start the "fortran" code:

```python
from subprocess import Popen
p = Popen(['./fortran-code', '5'])
p.poll()
```

Watch the data files, and plot them as they become available

```python
import glob
import os
import time
from IPython.display import clear_output

last_seen = None

def mtime(fname):
    return os.stat(fname).st_mtime

previous = None

tic = time.time()
while p.poll() is None: # while the subprocess is running
    time.sleep(1)
    print("elapsed: %.0f" % (time.time() - tic))
    fnames = glob.glob("data_*.dat")
    if not fnames:
        # no data files
        continue
    # get the most recently modified file
    latest = sorted(fnames, key=mtime)[-1]
    # if the most recent file is not the currently displayed plot,
    # clear the output and make a new plot
    if latest != previous:
        previous = latest
        clear_output(wait=True)
        plot_fname(latest)
    
print("done")  
```
