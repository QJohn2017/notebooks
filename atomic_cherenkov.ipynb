```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
#constants
n = 3         # v_p/c_s
nkx = 512
sidek = 25.6
deltak = 0.1
sider = np.pi/deltak
deltar = np.pi/sidek
x = np.arange(-sider, sider, deltar)
y = np.arange(-sider, sider, deltar)
X, Y = np.meshgrid(x, y)
kappa = 0.3   # in units of g|Psi_0|^2
gv = 0.01     # in units of g|Psi_0|^2
```

```python
# Update the matplotlib configuration parameters:
matplotlib.rcParams.update({'font.size': 16, 'font.family': 'serif'})
```

```python
def index_mom(mom):
    return int( np.floor( (mom + sidek) / deltak ) )
```

```python
#k plotting interval
kxl, kxr = -6, 6
kyl, kyr = -6, 6
idx_kxl, idx_kxr = index_mom(kxl), index_mom(kxr)
idx_kyl, idx_kyr = index_mom(kyl), index_mom(kyr)
```

```python
## calculate rayleigh scattering ring in momentum space ##
kx = np.arange(-sidek, sidek, deltak)
ky = np.arange(-sidek, sidek, deltak)
KX, KY = np.meshgrid(kx, ky)
```

```python
KSQ = KX**2 + KY**2
Kn = n* KX
```

```python
NUMERATOR = KSQ/2 - Kn + 1.j * kappa
DENOMINATOR = (KSQ/2) * (KSQ/2 + 2) - (Kn - 1.j * kappa)**2
```

```python
dPSI = -gv*np.divide(NUMERATOR, DENOMINATOR)
APSI = np.log10(np.abs(dPSI)**2) #logscale
```

```python
#add constant part to wavefunction at k=0
dPSI[nkx/2, nkx/2] += nkx
```

```python
#compute fourier transform
PSIR = np.fft.fftshift(np.fft.ifft2(nkx * dPSI))
APSIR = np.abs(PSIR)**2 #dimensionless
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 14))

axes[0].imshow(APSI[idx_kyl:idx_kyr,idx_kxl:idx_kxr],\
               cmap=cm.gray, origin=None, extent=[kx[idx_kxl], kx[idx_kxr], ky[idx_kyl], ky[idx_kyr]])
axes[0].set_title(r'$\left|\tilde{\psi}\left(k+k_0\right)\right|^2 [(g_v^2/g) / (g |\psi_0|^2)]$')

axes[1].imshow(APSIR,\
               cmap=cm.gray, origin=None, extent=[x[0], x[-1], y[0], y[-1]])
axes[1].set_title(r'$I$')

axes[0].set_ylabel(r'$k_y [m c_s]$') 
axes[0].set_xlabel(r'$k_x [m c_s]$')

axes[1].set_ylabel(r'$y [(m c_s)^{-1}]$')    
axes[1].set_xlabel(r'$x [(m c_s)^{-1}]$')
```
