```python
import numpy as np
```

```python
#side_cut = 32.2
side_cut = 2 * np.pi
n_cut = 50
n_points = 100
x_cut = np.linspace(0, side_cut, n_cut, endpoint=False)
#k = 2 * np.pi / 8
k = 1
#print k
phi = 2 * np.pi / 3
```

```python
delta_cut = x_cut[1] - x_cut[0]
k_cut = 2 * np.pi * np.fft.rfftfreq(n_points, d=delta_cut)
```

```python
wave = 1 * np.cos(k * x_cut + phi)
wave_fft = np.fft.rfft(wave, n_points)
ampl_spect = np.abs(wave_fft) * (2./n_cut)
phase_spect = np.angle(wave_fft)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))

axes[0].set_title('cosine wave')
axes[0].plot(x_cut, wave, '-o')
axes[0].set_ylabel('amplitude')
axes[0].set_xlabel(r'$r [\mu m]$')

axes[1].set_title('amplitude spectrum')
axes[1].plot(k_cut, ampl_spect, '--o')
axes[1].set_ylabel('magnitude')
axes[1].set_xlabel(r'$k [\mu m^{-1}]$')

axes[2].set_title('phase spectrum')
axes[2].plot(k_cut, phase_spect, 'o')
axes[2].set_ylabel('phase [rad]')
axes[2].set_xlabel(r'$k [\mu m^{-1}]$')
```
