```julia
blp = Float64[OnePump.bsenlp(0., momx, n_p) for momx in kx];
gma = Float64[OnePump.γ(0., momx) for momx in kx]/2;
```

```julia
fig, ax = plt.subplots(figsize=(5, 3))

ax[:plot](kx, blp, "k-.")
ax[:fill_between](kx, blp - gma/2, blp + gma/2, alpha=0.5)

ax[:scatter](OnePump.kpx, OnePump.ωp)
    
ax[:set_title]("blue-shifted LP dispersion")
ax[:set_xlim](-8, 8)
ax[:set_xlabel](L"$k_x[\mu m^{-1}]$")
ax[:set_ylabel](L"$\epsilon-\omega_X[\gamma_p]$")
```

$$V_{d}(x, y)=\frac{g_{V}}{2\pi\sigma}\exp\left[-\frac{1}{2}\left(\frac{x-x_{0}}{\sigma}\right)^{2}\right]\exp\left[-\frac{1}{2}\left(\frac{y-y_{0}}{\sigma}\right)^{2}\right]$$

```julia
vd(y::Float64, x::Float64, σ::Float64, gv::Float64, y0::Float64, x0::Float64) = gv/(2pi*σ)*exp(-1/(2σ^2)*((x-x0)^2+(y-y0)^2))
```

```julia
vdmat = [vd(y, x, 1., 0.5, 10., -10.) for x=-20:0.25:20, y=-20:0.25:20];
```

```julia
fig, ax = plt.subplots(figsize=(4, 4))
ax[:imshow](vdmat', ColorMap("gray"), origin="lower",
                      extent=[-20, 20, -20, 20])
ax[:set_title]("defect potential")
ax[:set_ylim](-20, 20)
ax[:set_xlim](-20, 20)
ax[:set_xlabel](L"$x [\mu m]$")
ax[:set_ylabel](L"$y [\mu m]$")
```

```julia
ψmom = Complex{Float64}[OnePump.ψtmom(momy, momx, n_p, 0.2, 0.5, 10., -10.) for momx in kx, momy in kx];
```

```julia
#add constant part to wavefunction at k = 0
ψmom[257, 257] += 512/OnePump.hopfx(OnePump.kpy, OnePump.kpx);
```

```julia
fig_mom_P, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[1][:imshow](log10(n_p*abs2(ψmom')),
ColorMap("gray"), origin=None, extent=[kx[1], kx[end], kx[1], kx[end]])
axes[2][:plot](kx, log10(n_p*abs2(ψmom[:, 257]))) 
axes[2][:grid]()
axes[1][:set_ylabel](L"$k_y [\mu m^{-1}]$")
axes[1][:set_xlabel](L"$k_x [\mu m^{-1}]$")
axes[2][:set_xlabel](L"$k_x [\mu m^{-1}]$")
```

$$I(r)=\frac{\vert\psi(r)\vert^{2}}{\vert\psi_p^{\text{ss}}\vert^{2}}=\left|X_{p}\right|^{2}\left|\sum_{k}\left[\frac{\delta(k)}{X_{p}}+\frac{Q\left(k\right)R^{*}\left(-k\right)V_{d}^{*}\left(-k\right)-M^{*}\left(-k\right)R\left(k\right)V_{d}\left(k\right)}{M\left(k\right)M^{*}\left(-k\right)-Q\left(k\right)Q^{*}\left(-k\right)}\right]e^{ikr}\right|^{2}$$

```julia
ψr = abs2(OnePump.hopfx(OnePump.kpy, OnePump.kpx)*fftshift(ifft(512ψmom)));
```

```julia
fig_real_P, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[1][:imshow](ψr', ColorMap("gray"),
                 origin=None)
axes[2][:plot](ψr[:, 257])
axes[2][:grid]()
axes[1][:set_ylabel](L"$y [\mu m]$")
axes[1][:set_xlabel](L"$x [\mu m]$")
axes[2][:set_xlabel](L"$x [\mu m]$")
```
