```julia
# potential in real space
vd(y::Float64, x::Float64; σ=1., gv=OnePump.gv, y0=0., x0=0., a=1., b=1., α=0.) =
gv/(2pi*a*b*σ^2)*exp(-1/2σ^2*(((x-x0)*cos(α) + (y-y0)*sin(α))^2/a^2 + ((x-x0)*sin(α) - (y-y0)*cos(α))^2/b^2))
```

```julia
# potential matrix in r
σ = .05
a = 80.
b = 20.
α = torad(90.)
vdmat = Float64[vd(y, x; x0=xdef, y0=ydef, a=a, α=α, σ=σ, b=b) for x in sbox.x, y in sbox.y];
# pot mat in k
vdtmat = Complex{Float64}[vdt(ky, kx; x0=xdef, y0=ydef, a=a, α=α, σ=σ) for kx in sbox.kx, ky in sbox.ky];
```

```julia
# cropped pot mat in r
cvdmat, cx, cy = cropmat(transpose(vdmat), sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=35.);
# cropped pot mat in k
cvdtmat, ckx, cky = cropmat(transpose(vdtmat), sbox.kx, sbox.ky; xcenter=0., ycenter=0., side=2.);

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[1][:imshow](cvdmat, ColorMap("gray"), origin="lower", 
                     extent=[cx[1], cx[end], cy[1], cy[end]])
axes[2][:imshow](abs2(cvdtmat), ColorMap("gray"), 
                 extent=[ckx[1], ckx[end], cky[1], cky[end]])

axes[2][:set_title](L"$V_d(k)[\gamma_p \mu m^2]$")
axes[1][:set_title](L"$V_d(r)[\gamma_p]$")

axes[1][:set_xlabel](L"$x [\mu m]$")
axes[1][:set_ylabel](L"$y [\mu m]$")

axes[2][:set_xlabel](L"$k_x [\mu m^{-1}]$")
axes[2][:set_ylabel](L"$k_y [\mu m^{-1}]$")
```

```julia
# calculate wf in mom space
ψmom = Complex{Float64}[ψtmom(momy, momx; ωp=-33., np=20., x0=xdef, y0=ydef, σ=σ, a=a, α=α, b=b) for momx in sbox.kx, momy in sbox.ky];
```

```julia
#add constant part to wavefunction at k = 0
ψmom[nx/2+1, ny/2+1] += sqrt(nx*ny)/xp;
```

```julia
# crop
cψmom, ckx, cky = cropmat(transpose(ψmom), sbox.kx, sbox.ky; xcenter=0., ycenter=0., side=1.4);
```

```julia
# logscale, for plotting
cψmomplt = log10(20.*abs2(cψmom));
```

```julia
fig_mom_P, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[1][:imshow](cψmomplt, ColorMap("gray"), 
                 extent=[ckx[1], ckx[end], cky[1], cky[end]])
axes[1][:set_title](L"$g\left|\widetilde{\psi}\left(k+k_{p}\right)\right|^{2}[\gamma_p \mu m^4]$")
axes[2][:plot](ckx, vec(cψmomplt[div(size(cψmomplt)[1],2), :])) 
axes[2][:grid]()
axes[1][:set_ylabel](L"$k_y [\mu m^{-1}]$")
axes[1][:set_xlabel](L"$k_x [\mu m^{-1}]$")
axes[2][:set_xlabel](L"$k_x [\mu m^{-1}]$")
```

```julia
ψr = abs2(xp*fftshift(ifft(sqrt(nx*ny)ψmom)));
```

```julia
cψr, cx, cy = cropmat(transpose(ψr), sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=35.);
```
