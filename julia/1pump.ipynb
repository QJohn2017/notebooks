```julia
# exciton-exciton contact interaction in γp*μm^2  
g = 5e-6/OnePump.γp
# experimental blue-shift, in units of γp
bshift = 0.0012924472535611464/OnePump.γp
#global const n_p = bshift/2abs2(OnePump.hopfx(OnePump.kpy, OnePump.kpx))
global const n_p = 43.
# alternatively
#global const n_p = g*abs2(OnePump.hopfx(OnePump.kpy, OnePump.kpx))*650
```

$$n_{p}=g\left|X_{p}\right|^{2}\left|\psi_{p}^{\text{ss}}\right|^{2}$$

```julia
# inferred exp pump polariton density, in μm^-2
n_p/(g*abs2(OnePump.hopfx(OnePump.kpy, OnePump.kpx)))
```

```julia
# Computational parameters
global const kx = linspace(-12.8, 12.75, 512);
```

```julia
blp = Float64[OnePump.bsenlp(0., momx, n_p) for momx in kx];
gma = Float64[OnePump.γ(0., momx) for momx in kx]/2;
```

```julia
fig, ax = plt.subplots(figsize=(5, 3))

ax[:plot](kx, blp, "k-.")
ax[:fill_between](kx, blp - gma/2, blp + gma/2, alpha=0.5)

ax[:scatter](OnePump.kpx, OnePump.ωp)
    
ax[:set_title]("blue-shifted LP dispersion")
ax[:set_xlim](-8, 8)
ax[:set_xlabel](L"$k_x[\mu m^{-1}]$")
ax[:set_ylabel](L"$\epsilon-\omega_X[\gamma_p]$")
```

```julia
np = n_p
fig, axes = plt.subplots(2,1, figsize=(8,6))
for i = 1:2
    axes[1][:scatter](kx, real([OnePump.λ1(0., momx, np) for momx in kx]), s=15, alpha=0.4, color="orange")
    axes[1][:scatter](kx, real([OnePump.λ2(0., momx, np) for momx in kx]), s=15, alpha=0.2)
    axes[2][:scatter](kx, imag([OnePump.λ1(0., momx, np) for momx in kx]), s=15, alpha=0.4, color="orange")
    axes[2][:scatter](kx, imag([OnePump.λ2(0., momx, np) for momx in kx]), s=15, alpha=0.2)
end
for ax in axes
    ax[:set_xlim](kx[1], kx[end])
    ax[:grid]()
end
axes[1][:set_ylabel](L"$\Re(\omega)[\gamma_p]$")
axes[2][:set_ylabel](L"$\Im(\omega)[\gamma_p]$")
axes[2][:set_xlabel](L"$k_x[\mu m^{-1}]$")
```

```julia
global const ipchosen = OnePump.findpump(OnePump.kpy, OnePump.kpx, OnePump.ωp, n_p)
```

```julia
fig, ax = plt.subplots(figsize=(8, 3))
ipmax=2ipchosen
for ip in linspace(0, ipmax, 100)
    ips, nps, colors = OnePump.mfroots(OnePump.kpy, OnePump.kpx, OnePump.ωp, ip)
    for idx = 1:length(ips)
        ax[:scatter](ips[idx], nps[idx], s=15, alpha=0.4, color=colors[idx])
    end
end
ax[:axvline](x=ipchosen, color="black", ls="dashed")
ax[:grid]()
ax[:set_ylim](0, 60)
ax[:set_xlim](0, ipmax)
ax[:set_xlabel](L"$I_p [\gamma_p^3]$")
ax[:set_ylabel](L"$n_p [\gamma_p]$")
```

```julia
ψmom = Complex{Float64}[OnePump.ψtmom(momy, momx, n_p, 0.2, 0.5) for momx in kx, momy in kx];
```

```julia
#add constant part to wavefunction at k = 0
ψmom[257, 257] += 512/OnePump.hopfx(OnePump.kpy, OnePump.kpx);
```

```julia
fig_mom_P, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[1][:imshow](log10(n_p*abs2(ψmom')),
ColorMap("gray"), origin=None, extent=[kx[1], kx[end], kx[1], kx[end]])
axes[2][:plot](kx, log10(n_p*abs2(ψmom[:, 257]))) 
axes[2][:grid]()
axes[1][:set_ylabel](L"$k_y [\mu m^{-1}]$")
axes[1][:set_xlabel](L"$k_x [\mu m^{-1}]$")
axes[2][:set_xlabel](L"$k_x [\mu m^{-1}]$")
```
