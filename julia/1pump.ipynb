```julia
using Autoreload
arequire("OnePump")
arequire("Various")
using PyPlot
using Contour
```

```julia
const nx = ny = 1024;
const xdef, ydef = (-11.2,-0.2);
```

```julia
sider(;l=70., n=256) = Float64[-l+(iy-1)*(2l/n) for iy=1:n]
sidek(;l=70., n=256) = 2pi*fftshift(Various.fftfreq(n, 2l/n))
```

```julia
sbox = Various.sysbox{Float64}(map(Various.vec2range, (sider(;l=200., n=ny), sider(;l=200., n=nx), 
                                                       sidek(;l=200., n=ny), sidek(;l=200., n=nx)))...);
```

```julia
# exciton-exciton contact interaction in γp*μm^2  
g = 5e-6/OnePump.γp
# experimental blue-shift, in units of γp
bshift = 0.0012924472535611464/OnePump.γp
#global const n_p = bshift/2abs2(OnePump.hopfx(OnePump.kpy, OnePump.kpx))
const ρp = 43. 
# alternatively
#global const n_p = g*abs2(OnePump.hopfx(OnePump.kpy, OnePump.kpx))*650
```

$$n_{p}=g\left|X_{p}\right|^{2}\left|\psi_{p}^{\text{ss}}\right|^{2}$$

```julia
# inferred exp pump polariton density, in μm^-2
ρp/(g*abs2(OnePump.hopfx(OnePump.kpy, OnePump.kpx)))
```

```julia
blp = Float64[OnePump.bsenlp(0., momx, ρp) for momx in sbox.kx];
gma = Float64[OnePump.γ(0., momx) for momx in sbox.kx]/2;
```

```julia
fig, ax = plt.subplots(figsize=(5, 3))

ax[:plot](sbox.kx, blp, "k-.")
ax[:fill_between](sbox.kx, blp - gma/2, blp + gma/2, alpha=0.5)

ax[:scatter](OnePump.kpx, OnePump.ωp)
    
ax[:set_title]("blue-shifted LP dispersion")
ax[:set_xlim](sbox.kx[1], sbox.kx[end])
ax[:set_xlabel](L"$k_x[\mu m^{-1}]$")
ax[:set_ylabel](L"$\epsilon-\omega_X[\gamma_p]$")
```

```julia
function scatter(axes, axno, reim::Function, λno::Symbol, col; np=ρp)
    axes[axno][:scatter](sbox.kx, reim([OnePump.(λno)(0., momx, np) for momx in sbox.kx]), s=15, alpha=0.2, color=col)
end
```

```julia
fig, axes = plt.subplots(2,1, figsize=(8,6))

scatter(axes, 1, real, :λ1, "orange")
scatter(axes, 1, real, :λ2, "blue")
scatter(axes, 2, imag, :λ1, "orange")
scatter(axes, 2, imag, :λ2, "blue")

for ax in axes
    ax[:set_xlim](sbox.kx[1], sbox.kx[end])
    ax[:grid]()
end
axes[1][:set_ylabel](L"$\Re(\omega)[\gamma_p]$")
axes[2][:set_ylabel](L"$\Im(\omega)[\gamma_p]$")
axes[2][:set_xlabel](L"$k_x[\mu m^{-1}]$")
```

$$V_{d}(x, y)=\frac{g_{V}}{2\pi\sigma}\exp\left[-\frac{1}{2}\left(\frac{x-x_{0}}{\sigma}\right)^{2}\right]\exp\left[-\frac{1}{2}\left(\frac{y-y_{0}}{\sigma}\right)^{2}\right]$$

```julia
vd(y::Float64, x::Float64; σ=1., gv=OnePump.gv, y0=ydef, x0=xdef) = gv/(2pi*σ)*exp(-1/(2σ^2)*((x-x0)^2+(y-y0)^2))
```

```julia
vdmat = [vd(y, x; x0 = 100., y0=0.) for x in sbox.x, y in sbox.y];
```

```julia
cvdmat, cx, cy = Various.cropmat(vdmat', sbox.x, sbox.y; xcenter=100., ycenter=0., side=35.);
```

```julia
fig, ax = plt.subplots(figsize=(4, 4))
ax[:imshow](cvdmat, ColorMap("gray"), origin="lower", 
                     extent=[cx[1], cx[end], cy[1], cy[end]])
ax[:set_title](L"$V_d(r)[\gamma_p]$")
ax[:set_ylim](cy[1], cy[end])
ax[:set_xlim](cx[1], cx[end])
ax[:set_xlabel](L"$x [\mu m]$")
ax[:set_ylabel](L"$y [\mu m]$")
```

```julia
ψmom = Complex{Float64}[OnePump.ψtmom(momy, momx, ρp; x0=xdef, y0=ydef) for momx in sbox.kx, momy in sbox.ky];
```

```julia
#add constant part to wavefunction at k = 0
ψmom[ny/2+1, nx/2+1] += sqrt(nx*ny)/OnePump.hopfx(OnePump.kpy, OnePump.kpx);
```

```julia
cψmom, ckx, cky = Various.cropmat(ψmom', sbox.kx, sbox.ky; xcenter=0., ycenter=0., side=1.5);
```

```julia
# logscale, for plotting
cψmomplt = log10(ρp*abs2(cψmom));
```

```julia
fig_mom_P, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[1][:imshow](cψmomplt, ColorMap("gray"), 
                 extent=[ckx[1], ckx[end], cky[1], cky[end]])
axes[1][:set_title](L"$g\left|\widetilde{\psi}\left(k+k_{p}\right)\right|^{2}[\gamma_p \mu m^4]$")
axes[2][:plot](ckx, vec(cψmomplt[div(size(cψmomplt)[2],2), :])) 
axes[2][:grid]()
axes[1][:set_ylabel](L"$k_y [\mu m^{-1}]$")
axes[1][:set_xlabel](L"$k_x [\mu m^{-1}]$")
axes[2][:set_xlabel](L"$k_x [\mu m^{-1}]$")
```

$$I(r)=\frac{\vert\psi(r)\vert^{2}}{\vert\psi_p^{\text{ss}}\vert^{2}}=\left|X_{p}\sum_{k}\left[\frac{\delta(k)}{X_{p}}+\frac{Q\left(k\right)R^{*}\left(-k\right)V_{d}^{*}\left(-k\right)-M^{*}\left(-k\right)R\left(k\right)V_{d}\left(k\right)}{M\left(k\right)M^{*}\left(-k\right)-Q\left(k\right)Q^{*}\left(-k\right)}\right]e^{ikr}\right|^{2}$$

```julia
ψr = abs2(OnePump.hopfx(OnePump.kpy, OnePump.kpx)*fftshift(ifft(sqrt(nx*ny)ψmom)));
```

```julia
cψr, cx, cy = Various.cropmat(ψr', sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=35.);
```

```julia
parabolas = map(M -> Various.parabola([Various.parabx(M, t, xdef, 2*0.89) for t in -10:0.1:10],
                                      [Various.paraby(M, t, ydef) for t in -10:0.1:10]), 10:14);
```

```julia
fig_real_P, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[1][:imshow](cψr, ColorMap("gray"), extent=[cx[1], cx[end], cy[1], cy[end]])

for par in parabolas
    axes[1][:plot](par.xpoints, par.ypoints)
end
axes[1][:axis]("image")

axes[1][:set_ylim](cy[1], cy[end])
axes[1][:set_xlim](cx[1], cx[end])


axes[2][:plot](cx, vec(cψr[div(size(cψr)[2],2), :]))
axes[2][:grid]()
axes[1][:set_ylabel](L"$y [\mu m]$")
axes[1][:set_xlabel](L"$x [\mu m]$")
axes[2][:set_xlabel](L"$x [\mu m]$")
```
