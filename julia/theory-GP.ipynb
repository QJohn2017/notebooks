```julia
# rabi splitting and bottom of photon band, in eV 
ΩR = 4.4/2*1e-3 
ωC0 = ωX = 1.53;
```

$\omega_{C}(k)=\omega_{C}^0\,\sqrt{1+{k^2}/{k_z^2}}=\omega_{C}^{0}+\frac{\omega_{C}^{0}}{2k_{z}^{2}}k^{2}+\dots$

$k_{z}=\frac{1}{l_C}\sqrt{\frac{\omega_{C}^{0}}{2\Omega_{R}}}$

```julia
#photon wv. along growth (z) direction, in  μm^(-1)
kz = 1/lC*sqrt(ωC0/2ΩR)
println(kz)  #~20.0
```

```julia
enC(kx,ky) = ωC0 * sqrt(1 + (sqrt(kx^2 + ky^2) / kz)^2)
```

\begin{equation}
\omega_{UP,LP}(k)=\frac{\omega_C(k)+\omega_X(k)}{2} \pm\left[\Omega_R^2 + \left(\frac{\omega_C(k)-\omega_X(k)}{2}\right)^2 \right]^{1/2}
\end{equation}

```julia
enLP(kx,ky) = (enC(kx,ky) + ωX)/2 - sqrt(ΩR^2 + ((enC(kx,ky)-ωX)/2)^2)
```

```julia
enUP(kx,ky) = (enC(kx,ky) + ωX)/2 + sqrt(ΩR^2 + ((enC(kx,ky)- ωX)/2)^2)
```

```julia
#bottom of (unshifted) LP/UP, in eV
en_LPB = enLP(0,0)
en_UPB = enUP(0,0)
println(en_UPB - en_LPB - 2ΩR) #0
```

```julia
hopf_x(kx,ky) = 1 / sqrt(1 + (ΩR/(enLP(kx,ky) - enC(kx,ky)))^2)
```

```julia
#linewidths for exciton and photon, in eV
γX = 2*0.12ΩR
γC = 2*0.12ΩR;
```

```julia
γ(kx,ky) = γC + hopf_x(kx,ky)^2 * (γX - γC)
```

```julia
#linewidth for pump state, in eV
γp = γ(k_pmpx, k_pmpy)
```

$2\omega_{LP}(k_{p})=\omega_{LP}(k_{s})+\omega_{LP}(\left|2k_{p}-k_{s}\right|)$

```julia
#checking energy conservation
println(enLP(k_sigx,k_sigy) + enLP(2k_pmpx-k_sigx,2k_pmpy-k_sigy) - 2enLP(k_pmpx,k_pmpy))  #0
```
