```julia
const lx = 100.
const ly = 100.
sbox = sysbox{Float64}(sider(;l=ly, n=ny), sider(;l=lx, n=nx), sidek(;l=ly, n=ny), sidek(;l=lx, n=nx));
```

```julia
vp = 1.5
t = linspace(0., 2pi, 45);
xcirc = vp + vp*cos(t)
ycirc = vp*sin(t);
```

```julia
function ψt(ky::Float64, kx::Float64; vp=1.5, Δp=0., κ=0.1)
    numer = -kx*vp + ε(ky,kx; Δp=Δp) + im*κ
    denom = λ1(ky, kx; Δp=Δp, vp=(0., vp), κ=κ) * λ2(ky, kx; Δp=Δp, vp=(0., vp), κ=κ)
    numer/denom
end
```

```julia
ψtmat = Complex{Float64}[ψt(momy, momx) for momx in sbox.kx, momy in sbox.ky];
```

```julia
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

axes[1][:imshow](abs2(transpose(ψtmat)), ColorMap("gray"), 
extent=[sbox.kx[1], sbox.kx[end], sbox.ky[1], sbox.ky[end]])
axes[1][:set_title](L"$\left|\widetilde{\psi}\left(k+k_{p}\right)\right|^2$")

axes[2][:imshow](log10(abs2(transpose(ψtmat))), ColorMap("gray"), 
extent=[sbox.kx[1], sbox.kx[end], sbox.ky[1], sbox.ky[end]])
axes[2][:set_title](L"$\log\left|\widetilde{\psi}\left(k+k_{p}\right)\right|^2$")

axes[2][:scatter](xcirc, ycirc, s=4, alpha=0.2, color="orange")

ax[:set_xlim](-3.1,3.1)
ax[:set_ylim](-1.7,1.7)

axes[1][:set_ylabel](L"$\delta k_y[m c_s]$")

for ax in axes
    ax[:set_xlabel](L"$\delta k_x[m c_s]$")
    
    ax[:set_xlim](-3.1,3.1)
    ax[:set_ylim](-1.7,1.7)
end
```

```julia
function ψt1(ky::Float64, kx::Float64; vp=1.5, Δp=0., κ=0.1)
    numer = -kx*vp +ε(ky, kx; Δp=Δp) #+im*κ
    denom = λ1(ky, kx; Δp=Δp, vp=(0., vp), κ=κ)
    numer/denom
end
```

$$\frac{\epsilon^{2}(k)}{\left|\omega^{+}(k)\right|^{2}}-2\frac{\epsilon(k)kv_{p}}{\left|\omega^{+}(k)\right|^{2}}+\frac{\left(kv_{p}\right)^{2}}{\left|\omega^{+}(k)\right|^{2}}$$

$$\left(\frac{\epsilon(k)-kv_{p}}{\left|\omega^{+}(k)\right|}\right)^{2}$$

```julia
function ψt11(ky::Float64, kx::Float64; vp=1.5, Δp=0., κ=0.1)
    numer = ε(ky, kx; Δp=Δp)^2
    denom = λ1(ky, kx; Δp=Δp, vp=(0., vp), κ=κ)
    numer/abs2(denom)
end
```

```julia
function ψt12(ky::Float64, kx::Float64; vp=1.5, Δp=0., κ=0.1)
    numer = -2ε(ky, kx; Δp=Δp)*kx*vp
    denom = λ1(ky, kx; Δp=Δp, vp=(0., vp), κ=κ)
    numer/abs2(denom)
end
```

```julia
function ψt13(ky::Float64, kx::Float64; vp=1.5, Δp=0., κ=0.1)
    numer = (kx*vp)^2
    denom = λ1(ky, kx; Δp=Δp, vp=(0., vp), κ=κ)
    numer/abs2(denom)
end
```

```julia
λpls(ky::Float64, kx::Float64; vp=(0., 1.9), Δp=0., κ=1.1) = (ky*vp[1] + kx*vp[2]) + sqrt(ε(ky, kx; Δp=Δp)*(ε(ky, kx; Δp=Δp) + 2)) #-im*κ
```

```julia
function ψt4(ky::Float64, kx::Float64; vp=1.5, Δp=0., κ=0.1)
    numer = ε(ky, kx; Δp=Δp) - kx*vp
    denom = λpls(ky, kx; Δp=Δp, vp=(0., vp), κ=κ)
    (numer/denom)^2
end
```
