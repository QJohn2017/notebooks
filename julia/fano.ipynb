```julia
using OnePumpQuad
using Various
using PyPlot
```

```julia
const nx = ny = 512;
```

```julia
const lx = 100.
const ly = 100.
sbox = sysbox{Float64}(sider(;l=ly, n=ny), sider(;l=lx, n=nx), sidek(;l=ly, n=ny), sidek(;l=lx, n=nx));
```

```julia
function scatter(axes, axno, reim::Function, λ::Function, col)
    axes[axno][:scatter](sbox.kx, reim([λ(0., momx; vp=(0.,1.5), κ=0.1, Δp=0.) for momx in sbox.kx]), s=15, alpha=0.5, color=col)
end
```

```julia
fig, axes = plt.subplots(2,1, figsize=(8,6))

scatter(axes, 1, real, λ1, "orange")
scatter(axes, 1, real, λ2, "blue")
scatter(axes, 2, imag, λ1, "orange")
scatter(axes, 2, imag, λ2, "blue")

for ax in axes
    ax[:set_xlim](-2.5, 2.5)
    ax[:grid]()
end

axes[1][:set_ylim](-4, 4)
axes[2][:set_ylim](-2, 0)

axes[1][:set_ylabel](L"$\Re(\omega)[g \vert\psi_0\vert^2]$")
axes[2][:set_ylabel](L"$\Im(\omega)[g \vert\psi_0\vert^2]$")
axes[2][:set_xlabel](L"$\delta k_x[m c_s]$")
```

```julia
function ψt(ky::Float64, kx::Float64; vp=1.5, Δp=0., κ=0.1)
    numer = ε(ky, kx; Δp=Δp) -kx*vp +im*κ
    denom = λ1(ky, kx; Δp=Δp, vp=(0., vp), κ=κ) * λ2(ky, kx; Δp=Δp, vp=(0., vp), κ=κ)
    numer/denom
end
```

```julia
ψtmat = Complex{Float64}[ψt(momy, momx) for momx in sbox.kx, momy in sbox.ky];
```

```julia
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

axes[1][:imshow](abs2(transpose(ψtmat)), ColorMap("gray"), 
extent=[sbox.kx[1], sbox.kx[end], sbox.ky[1], sbox.ky[end]])
axes[1][:set_title](L"$\left|\widetilde{\psi}\left(k+k_{p}\right)\right|^2$")

axes[2][:imshow](log10(abs2(transpose(ψtmat))), ColorMap("gray"), 
extent=[sbox.kx[1], sbox.kx[end], sbox.ky[1], sbox.ky[end]])
axes[2][:set_title](L"$\log\left|\widetilde{\psi}\left(k+k_{p}\right)\right|^2$")

for ax in axes
    ax[:set_xlabel](L"$\delta k_x[m c_s]$")
    ax[:set_ylabel](L"$\delta k_y[m c_s]$")
end
```
