**Response function**

$$\det\left(\mathcal{L}\right)=D(q,\omega)=-\left(\omega-\omega_{+}(q)\right)\left(\omega-\omega_{-}(q)\right)$$

$$\chi(q,\omega)=-\frac{1}{D(q,\omega)}\left\{ X_{p}\left[\left(M^{*}(-q)+\omega\right)R(q)-Q(q)R^{*}(-q)\right]+X_{p}^{*}\left[\left(M(q)-\omega\right)R^{*}(-q)-Q^{*}(-q)R(q)\right]\right\}$$

Let's plot the response function for a static defect ($V=0$).

```julia
# calculate response function
χmat = Complex{Float64}[χ(momy, momx; ωp=-33., V=[0., 0.]) for momx in sbox.kx, momy in sbox.ky];
```

```julia
# crop
cχmat, ckx, cky = cropmat(transpose(χmat), sbox.kx, sbox.ky; xcenter=0., ycenter=0., side=2.);
```

```julia
# plot
fig_response, axes = plt.subplots(3, 2, figsize=(10, 14))
axes[1, 1][:imshow](real(cχmat), ColorMap("gray"), 
                 extent=[ckx[1], ckx[end], cky[1], cky[end]])
axes[1, 1][:set_title](L"$\Re\left(\chi(q, 0)\right)$")
axes[1, 2][:plot](ckx, vec(real(cχmat[div(size(cχmat)[1],2), :])))
axes[1, 2][:set_ylabel](L"$\Re\left(\chi(q, 0)\right)$")

axes[2, 1][:imshow](imag(cχmat), ColorMap("gray"), 
                 extent=[ckx[1], ckx[end], cky[1], cky[end]])
axes[2, 1][:set_title](L"$\Im\left(\chi(q, 0)\right)$")
axes[2, 2][:plot](ckx, vec(imag(cχmat[div(size(cχmat)[1],2), :])))
axes[2, 2][:set_ylabel](L"$\Im\left(\chi(q, 0)\right)$")

axes[3, 1][:imshow](log10(abs(cχmat)), ColorMap("gray"), 
                 extent=[ckx[1], ckx[end], cky[1], cky[end]])
axes[3, 1][:set_title](L"$\vert\chi(q, 0)\vert$")
axes[3, 2][:plot](ckx, vec(abs(cχmat[div(size(cχmat)[1],2), :])))
axes[3, 2][:set_ylabel](L"$\vert\chi(q, 0)\vert$")

for ax in axes[:, 2]
    ax[:grid]()
    ax[:set_xlim](ckx[1], ckx[end])
end

for ax in axes[:, 1]
    ax[:set_ylabel](L"$q_y [\mu m^{-1}]$")
end

for ax in axes
   ax[:set_xlabel](L"$q_x [\mu m^{-1}]$") 
end
```

**Real-space density perturbation**

$$\delta\rho(q,-Vq)=\chi(q,-Vq)f_{d}(q)$$

```julia
# calculate density modulation in k-space
δρmat = Complex{Float64}[δρ(momy, momx; ωp=-33., gV=1., V=[0., 0.]) for momx in sbox.kx, momy in sbox.ky];
```

$$\delta\tilde{\rho}(\hat{r})=\int dqe^{iq\hat{r}}\delta\rho(q,-Vq)$$

$$\left|\tilde{\psi}(r,t)\right|^{2}=\frac{n_{p}}{g\left|X_{p}\right|^{2}}\left[1+\delta\tilde{\rho}(r,t)\right]$$

```julia
# ifft to obtain density modulation in real-space
δρtmat = real(fftshift(ifft(fftshift(sqrt(nx*ny)*δρmat))));
```

```julia
# crop
cδρtmat, cx, cy = cropmat(transpose(δρtmat), sbox.x, sbox.y; xcenter=0., ycenter=0., side=35.);
```

```julia
# plot
fig_real_density, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[1][:imshow](cδρtmat, ColorMap("gray"), extent=[cx[1], cx[end], cy[1], cy[end]])

axes[2][:plot](cx, vec(cδρtmat[div(size(cδρtmat)[1],2), :]))
axes[2][:set_xlim](cx[1], cx[end])
axes[2][:grid]()

axes[1][:set_ylabel](L"$y [\mu m]$")
axes[1][:set_xlabel](L"$x [\mu m]$")
axes[2][:set_ylabel](L"$I$")
axes[2][:set_xlabel](L"$x [\mu m]$")
```

**Drag force**

$$\tilde{f}_{d}(r)=\frac{g_{V}}{2\pi\sigma^{2}}\exp\left[-\frac{r^{2}}{2\sigma^{2}}\right]$$

$$\left(F_{d}\right)_{x}=-\frac{n_{p}}{g\left|X_{p}\right|^{2}}\frac{1}{\sigma^2}\frac{g_{V}}{2\pi\sigma^{2}}\int\mathrm{d}xdy\,\delta\tilde{\rho}(x,y)x\exp\left[-\frac{x^{2}+y^{2}}{2\sigma^{2}}\right]$$

$$\left(F_{d}\right)_{y}=-\frac{n_{p}}{g\left|X_{p}\right|^{2}}\frac{g_{V}}{2\pi\sigma^{4}}\int\mathrm{d}xdy\,\delta\tilde{\rho}(x,y)y\exp\left[-\frac{x^{2}+y^{2}}{2\sigma^{2}}\right]$$

```julia
# calculate r.V on grid
fdtmat = Float64[rx*fdt(ry, rx) for rx in sbox.x, ry in sbox.y];
norm(fdtmat)
```

```julia
-sum(fdtmat.*δρtmat)
```

```julia
function drag(box; ωp=-30., np=20., V=[0., 0.], gV=1., σ=1.)
    # calculate density modulation in k-space
    δρmat = Complex{Float64}[δρ(momy, momx; ωp=ωp, np=np, V=V, gV=gV, σ=σ) for momx in box.kx, momy in box.ky]
    # ifft to obtain density modulation in real-space
    N = length(box.kx)*length(box.ky)
    δρtmat = real(fftshift(ifft(fftshift(sqrt(N)*δρmat))))
    # calculate r.V on grid
    fdtm = Float64[rx*fdt(ry, rx; σ=σ, gV=gV) for rx in box.x, ry in box.y]
    @show norm(fdtm)
    -sum(fdtmat.*δρtmat)
end
```
