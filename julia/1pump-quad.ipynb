$$\widetilde{\psi}(k+k_{p})=g_{V}\frac{\epsilon(k)-kv_{p}+i\kappa}{\omega^{-}(k)\omega^{+}(k)}$$

```julia
function ψt(ky::Float64, kx::Float64; vp=1.9, Δp=0., κ=1.1)
    numer = ε(ky, kx; Δp=Δp) - kx*vp + im*κ
    denom = λ1(ky, kx; Δp=Δp, vp=(0., vp), κ=κ) * λ2(ky, kx; Δp=Δp, vp=(0., vp), κ=κ)
    numer/denom
end
```

$$\Re\left[\widetilde{\psi}(k+k_{p})\right]=-\frac{\left[\frac{k^{2}}{2}\left(\frac{k^{2}}{2}+2\right)-k^{2}v_{p}^{2}\right]\left(\frac{k^{2}}{2}-kv_{p}\right)+\kappa^{2}\left(\frac{k^{2}}{2}+kv_{p}\right)}{\left[\frac{k^{2}}{2}\left(\frac{k^{2}}{2}+2\right)-k^{2}v_{p}^{2}+\kappa^{2}\right]^{2}+4\kappa^{2}k^{2}v_{p}^{2}}$$

```julia
function rψtnum(k::Float64; vp=1.9, κ=1.1)
    (k^2/2*(k^2/2+2)-(k*vp)^2)*(k^2/2-k*vp) + κ^2*(k^2/2+k*vp)
end
```

```julia
function rψtdenom(k::Float64; vp=1.9, κ=1.1)
    (k^2/2*(k^2/2+2)-(k*vp)^2+κ^2)^2 + 4κ^2*(k*vp)^2
end
```

$$\chi(q)=-\frac{\epsilon(k_{p}-q)+\epsilon(k_{p}+q)-2\omega_{p}+2g\vert P\vert^{2}}{D(q,\omega)}=\frac{k^{2}}{\omega_{+}\omega_{-}}$$

```julia
χ(k::Float64; vp=1.9, κ=1.1) = k^2/(λ1(0., k; vp=(0.,vp), κ=κ, Δp=0.)*λ2(0., k; vp=(0.,vp), κ=κ, Δp=0.))
```
