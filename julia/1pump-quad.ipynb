```julia
using OnePumpQuad
using Various
using PyPlot
using Contour
```

```julia
const nx = ny = 512;
```

```julia
const lx = 100.
const ly = 100.
sbox = sysbox{Float64}(sider(;l=ly, n=ny), sider(;l=lx, n=nx), sidek(;l=ly, n=ny), sidek(;l=lx, n=nx));
```

```julia
function scatter(axes, axno, reim::Function, λ::Function, col)
    axes[axno][:scatter](sbox.kx, reim([λ(0., momx; vp=(0.,3), κ=0.3) for momx in sbox.kx]), s=15, alpha=0.5, color=col)
end
```

```julia
idxkr = getidx(3., maximum(sbox.kx), step(sbox.kx))
idxkl = getidx(-3., maximum(sbox.kx), step(sbox.kx))
#check
sbox.kx[idxkl] + 3. #0
```

```julia
fig, axes = plt.subplots(2,1, figsize=(8,6))

scatter(axes, 1, real, quadλ1, "orange")
scatter(axes, 1, real, quadλ2, "blue")
scatter(axes, 2, imag, quadλ1, "orange")
scatter(axes, 2, imag, quadλ2, "blue")

for ax in axes
    ax[:set_xlim](sbox.kx[idxkl], sbox.kx[idxkr])
    ax[:grid]()
end

axes[1][:set_ylim](-4, 4)
axes[2][:set_ylim](-2, 0)

axes[1][:set_ylabel](L"$\Re(\omega)[g \vert\psi_0\vert^2]$")
axes[2][:set_ylabel](L"$\Im(\omega)[g \vert\psi_0\vert^2]$")
axes[2][:set_xlabel](L"$\delta k_x[m c_s]$")
```

```julia
quadψmom = Complex{Float64}[quadψtmom(momy, momx; vp=(0., 3.), κ=0.3) for momx in sbox.kx, momy in sbox.ky];
```

```julia
#add constant part to wavefunction at k = 0
quadψmom[nx/2+1, ny/2+1] += sqrt(nx*ny);
```

```julia
# crop to interest region
quadcψmom, ckx, cky = cropmat(transpose(quadψmom), sbox.kx, sbox.ky; xcenter=3.1, ycenter=0., side=4.);
```

```julia
# logscale, for plotting
quadcψmomplt = log10(abs2(quadcψmom));
```

```julia
fig_mom_P, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[1][:imshow](quadcψmomplt, ColorMap("gray"), 
                 extent=[ckx[1], ckx[end], cky[1], cky[end]])
axes[1][:set_title](L"$\left|\widetilde{\psi}\left(k+k_{p}\right)\right|^{2}$")
axes[2][:plot](ckx, vec(quadcψmomplt[div(size(quadcψmomplt)[1],2), :])) 
axes[2][:grid]()
axes[1][:set_ylabel](L"$\delta k_y[m c_s]$")
axes[1][:set_xlabel](L"$\delta k_x[m c_s]$")
axes[2][:set_xlabel](L"$\delta k_x[m c_s]$")
```

```julia
# fft to real space
quadψr = abs2(fftshift(ifft(sqrt(nx*ny)quadψmom)));
```

```julia
#crop to interest region
quadcψr, cx, cy = cropmat(transpose(quadψr), sbox.x, sbox.y; side=20.);
```

```julia
fig_real_P, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[1][:imshow](quadcψr, ColorMap("gray"), extent=[cx[1], cx[end], cy[1], cy[end]])

axes[1][:set_ylim](cy[1], cy[end])
axes[1][:set_xlim](cx[1], cx[end])

axes[2][:plot](cx, vec(quadcψr[div(size(quadcψr)[1],2), :]))
axes[2][:grid]()
axes[1][:set_ylabel](L"$y [(m c_s)^{-1}]$")
axes[1][:set_xlabel](L"$x [(m c_s)^{-1}]$")
axes[2][:set_xlabel](L"$x [(m c_s)^{-1}]$")
```
