```julia
# System parameters
global const omega_X = omega_C0 = 1.4
global const kz = 20.
global const Omega_R = 2.2/10^3
global const gamma_X = gamma_C = 0.1/10^3
global const k_pmpx = 0.6
global const k_pmpy = 0.
global const gp = gamma_C + (1 / sqrt(1 + (Omega_R / ((0.5 * ((omega_C0 * sqrt(1 + (sqrt(k_pmpx^2 + k_pmpy^2) / kz)^2)) + omega_X) - 0.5 * sqrt(((omega_C0 * sqrt(1 + (sqrt(k_pmpx^2 + k_pmpy^2) / kz)^2)) - omega_X)^2 + 4Omega_R^2)) - (omega_C0 * sqrt(1 + (sqrt(k_pmpx^2 + k_pmpy^2) / kz)^2))))^2))^2 * (gamma_X - gamma_C)
```

```julia
enC(kx,ky) = (omega_C0 * sqrt(1. + (sqrt(kx^2 + ky^2) / kz)^2) - omega_X) / gp
enLP(kx,ky) = 0.5 * enC(kx,ky) - 0.5 * sqrt(enC(kx,ky)^2 + 4.Omega_R^2/gp^2)
hopf_x(kx,ky) = 1. / sqrt(1. + ((Omega_R/gp) / (enLP(kx,ky) - enC(kx,ky)))^2)
gamma(kx,ky) = (gamma_C + hopf_x(kx,ky)^2 * (gamma_X - gamma_C)) / gp
```

```julia
global const omega_p = enLP(k_pmpx, k_pmpy) + 2.
```

$$L=\begin{pmatrix}M(k) & Q(k)\\
-Q^{*}(-k) & -M^{*}(-k)
\end{pmatrix}$$
$$M(k)=\epsilon(k_{p}+k)-\omega_{p}-\frac{i}{2}\gamma(k_{p}+k)+2n_{p}\left|X(k_{p}+k)\right|^{2}$$
$$Q(k)=n_{p}e^{2i\phi_{p}}X^{*}(k_{p}+k)X^{*}(k_{p}-k)$$

```julia
M(kx, ky, np) = enLP(k_pmpx + kx, k_pmpy + ky) - omega_p - im*gamma(k_pmpx + kx, k_pmpy + ky) + 2np*abs2(hopf_x(k_pmpx + kx, k_pmpy + ky))
Q(kx, ky, np) = np*conj(hopf_x(k_pmpx + kx, k_pmpy + ky))*conj(hopf_x(k_pmpx - kx, k_pmpy - ky))
L(kx, ky, np) = [M(kx, ky, np) Q(kx, ky, np); -conj(Q(-kx, -ky, np)) -conj(M(-kx, -ky, np))]
```

```julia
# Computational parameters
global const kx = linspace(-12.8, 12.75, 512);
```

```julia
function evmat(klist, np)
    len = length(klist)
    mat = Array(Complex{Float64}, (2, len))
    for i = 1:len
        mat[:,i] = eigvals(L(klist[i], 0., np))
    end
    return mat
end
```

```julia
ev = evmat(kx, 10.);
```

```julia
for i = 1:2
    scatter(kx, real(ev[i,:]), s=15, alpha=0.4)
end
grid()
xlim(kx[1], kx[end])
xlabel(L"$k_x[\mu m^{-1}]$")
ylabel(L"$\Re(\omega)[\gamma_p]$")
```

```julia
for i=1:2
    scatter(kx, imag(ev[i,:]), s=15, alpha=0.4)
end
grid()
xlim(kx[1], kx[end])
xlabel(L"$k_x[\mu m^{-1}]$")
ylabel(L"$\Im(\omega)[\gamma_p]$")
```

$$\lambda_{r}\left(k\right)_{1,2}=\Re(w)\pm\frac{1}{2}\sqrt{\frac{|z|+\Re(z)}{2}}$$
$$\lambda_{i}\left(k\right)_{1,2}=\Im(w)\pm\frac{1}{2}\text{sgn}(\Im(z))\sqrt{\frac{|z|-\Re(z)}{2}}$$
$$w(k)=M\left(k\right)-M^{*}\left(-k\right)$$
$$z(k)=\left[M\left(k\right)+M^{*}\left(-k\right)\right]^{2}-4Q\left(k\right)Q^{*}\left(-k\right)$$

```julia
w(kx, ky, np) = M(kx, ky, np) - conj(M(-kx, -ky, np))
z(kx, ky, np) = (M(kx, ky, np) + conj(M(-kx, -ky, np)))^2 - 4*Q(kx, ky, np)*conj(Q(-kx, -ky, np))
```

```julia
λr1(kx, ky, np) = 1/2*real(w(kx, ky, np)) + 1/2*sqrt((abs(z(kx, ky, np)) + real(z(kx, ky, np)))/2)
λr2(kx, ky, np) = 1/2*real(w(kx, ky, np)) - 1/2*sqrt((abs(z(kx, ky, np)) + real(z(kx, ky, np)))/2)
λi1(kx, ky, np) = 1/2*imag(w(kx, ky, np)) + 1/2*sign(imag(z(kx, ky, np)))*sqrt((abs(z(kx, ky, np)) - real(z(kx, ky, np)))/2)
λi2(kx, ky, np) = 1/2*imag(w(kx, ky, np)) - 1/2*sign(imag(z(kx, ky, np)))*sqrt((abs(z(kx, ky, np)) - real(z(kx, ky, np)))/2)
```

```julia
scatter(kx, [λr1(k, 0., 10.) for k in kx], s=15, alpha=0.4)
scatter(kx, [λr2(k, 0., 10.) for k in kx], s=15, alpha=0.4)
grid()
xlim(kx[1], kx[end])
xlabel(L"$k_x[\mu m^{-1}]$")
ylabel(L"$\Re(\omega)[\gamma_p]$")
```

```julia
scatter(kx, [λi1(k, 0., 10.) for k in kx], s=15, alpha=0.4)
scatter(kx, [λi2(k, 0., 10.) for k in kx], s=15, alpha=0.4)
grid()
xlim(kx[1], kx[end])
xlabel(L"$k_x[\mu m^{-1}]$")
ylabel(L"$\Re(\omega)[\gamma_p]$")
```

```julia
function mfcoefs(kpx::Float64, kpy::Float64, ωp::Float64, Ip::Float64)
    xp = hopf_x(kpx, kpy)
    ep = enLP(kpx, kpy)
    a = abs2(xp)^2
    b = 2abs2(xp)*(ep - ωp)
    c = 1/4 + (ep - ωp)^2
    d = -a*Ip
    r = roots(Poly([a, b, c, d]))
    filter!(x -> isapprox(imag(x), 0), r)
    ips = Array(Float64, size(r))
    fill!(ips, Ip)
    return (ips, real(r))
end
```

```julia
for ip = 0:.1:10
    scatter(mfcoefs(k_pmpx, k_pmpy, omega_p, ip)..., s=15, alpha=0.4)
end
grid()
ylim(0, 5)
xlim(0, 10)
xlabel(L"$I_p [\gamma_p^3]$")
ylabel(L"$n_p [\gamma_p]$")
```
