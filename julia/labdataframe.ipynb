```julia
function doubleeach2(arr::Vector{Int})
    out = Int[]
    for i = 1:2:length(arr)
        tmp = [arr[i], arr[i+1]]
        append!(out, repeat(tmp, outer=[2]))
    end
    return out
end
```

```julia
tods = repeat(['M', 'T'], outer=[30]);
```

```julia
nov = [date(2014,11,day) for day in repeat([3,4,5,6,17,18,19,20], inner = [2])]
dic = [date(2014,12,day) for day in repeat([1,2,3,4], inner = [2])]
ene = [date(2015,1,day) for day in repeat([26,27,28,29], inner = [2])]
feb = [date(2015,2,day) for day in repeat([9,10,11,12,23,24,25,26], inner = [2])]
mar = [date(2015,3,day) for day in repeat([2,3,4,5,16,17], inner = [2])]
dates = vcat(nov, dic, ene, feb, mar);
```

```julia
grnov = doubleeach2([5164, 5114, 5165, 5115, 5164, 5114, 5165, 5115])
grdic = doubleeach2([5164, 5114, 5165, 5115])
grene = doubleeach2([5164, 5114, 5165, 5115])
grfeb = doubleeach2([5164, 5114, 5165, 5115, 5164, 5114, 5165, 5115])
grmar = vcat(doubleeach2([5164, 5114, 5165, 5115]), [5164, 5114, 5165, 5115]);
groups = vcat(grnov, grdic, grene, grfeb, grmar);
```

```julia
labf = DataFrame(Date = dates, Group = groups, TOD = tods);
```
