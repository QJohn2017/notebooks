```julia
# modules
using PyPlot
using Contour
using Images
using Reactive
using Interact
```

```julia
# my modules
using Various
using OnePumpExp
```

```julia
matplotlib["rcParams"][:update](["font.size" => 18, "font.family" => "serif"])
```

```julia
# exp params
const NY = 200
const NX = 1024
const Δy = 0.64
const Δx = 0.16

const y = sider(;l=(NY-1)*Δy/2, n=NY)
const x = sider(;l=(NX-1)*Δx/2, n=NX)

const xdef, ydef = -11.2, -0.2;
# read exp data
ctpmpdata, cexpx, cexpy = gettrunccrop("E-pump.dat", x, y; xcenter=xdef, ycenter=ydef)
ctidldata, cexpx, cexpy = gettrunccrop("E-idler.dat", x, y; xcenter=xdef, ycenter=ydef)
ctsigdata, cexpx, cexpy = gettrunccrop("E-signal.dat", x, y; xcenter=xdef, ycenter=ydef)
# theoretical params
const nx = 2*NX
const ny = 2*NY;
```

```julia
const lx = 2*maximum(x)
const ly = 2*maximum(y)
sbox = sysbox{Float64}(sider(;l=ly, n=ny), sider(;l=lx, n=nx), sidek(;l=ly, n=ny), sidek(;l=lx, n=nx));
```

```julia
# Gaussian filtering
sigma = [4, 4]
gaussctpmpdata = Images.imfilter_gaussian(ctpmpdata, sigma)
gaussctidldata = Images.imfilter_gaussian(ctidldata, sigma)
gaussctsigdata = Images.imfilter_gaussian(ctsigdata, sigma)
diffctpmpdata = ctpmpdata - gaussctpmpdata
diffctidldata = ctidldata - gaussctidldata;
diffctsigdata = ctsigdata - gaussctsigdata;
```

```julia
f = figure();
pmpen=slider(-40:1.:-20)
pmpdens=slider(0:1.:30)
replot = button("Replot") # Commit your changes
map(display, [pmpen, pmpdens, replot]) # optional
coeffs = sampleon(replot, lift(tuple, pmpen, pmpdens))
@manipulate for pmp=coeffs; withfig(f) do
        ψmom = Complex{Float64}[ψtmom(momy, momx; ωp=pmp[1], np=pmp[2], x0=xdef, y0=ydef) for momx in sbox.kx, momy in sbox.ky];
        ψmom[nx/2+1, ny/2+1] += sqrt(nx*ny)/xp;
        ψr = abs2(xp*fftshift(ifft(sqrt(nx*ny)ψmom)));
        cψr, cx, cy = cropmat(transpose(ψr), sbox.x, sbox.y; xcenter=xdef, ycenter=ydef);
        c = contour(cx, cy, transpose(cψr), 1.0);
        PyPlot.imshow(diffctidldata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])
        for line in c.lines
            xs, ys = coordinates(line)
            PyPlot.plot(xs, ys, "y-")
        end
        PyPlot.ylim(cexpy[1], cexpy[end])
        PyPlot.xlim(cexpx[1], cexpx[end])
        PyPlot.xlabel(L"$x\,[\mu m]$")
        PyPlot.ylabel(L"$y\,[\mu m]$")
        PyPlot.scatter(xdef, ydef, s=40, c="orange", marker="o")
    end
end
```
