$$n_{p}=\frac{k_{p}^{2}}{\left|X_{p}\right|^{2}m_{LP}}\sin^{2}\phi$$
$$\omega_{p}=\epsilon_{p}^{\prime}+n_{p}\left|X_{p}\right|^{2}$$

```julia
# exp params
const NY = 200
const NX = 1024
const Δy = 0.64
const Δx = 0.16

const y = sider(;l=(NY-1)*Δy/2, n=NY)
const x = sider(;l=(NX-1)*Δx/2, n=NX)

const xdef, ydef = -11.2, -0.2;
# read exp data
ctpmpdata, cexpx, cexpy = gettrunccrop("E-pump.dat", x, y; xcenter=xdef, ycenter=ydef)
ctidldata, cexpx, cexpy = gettrunccrop("E-idler.dat", x, y; xcenter=xdef, ycenter=ydef)
# theoretical params
const nx = 2*NX
const ny = 2*NY;
```

```julia
const lx = 2*maximum(x)
const ly = 2*maximum(y)
sbox = sysbox{Float64}(sider(;l=ly, n=ny), sider(;l=lx, n=nx), sidek(;l=ly, n=ny), sidek(;l=lx, n=nx));
```

```julia
# Gaussian filtering
sigma = [4, 4]
gaussctpmpdata = Images.imfilter_gaussian(ctpmpdata, sigma)
gaussctidldata = Images.imfilter_gaussian(ctidldata, sigma)
diffctpmpdata = ctpmpdata - gaussctpmpdata
diffctidldata = ctidldata - gaussctidldata;
```

```julia
# lin response in k space
# ωp=-33., np=20.
ψmom = Complex{Float64}[ψtmom(momy, momx; ωp=-33., np=20., x0=xdef, y0=ydef) for momx in sbox.kx, momy in sbox.ky];
```

```julia
#add constant part to wavefunction at k = 0
ψmom[nx/2+1, ny/2+1] += sqrt(nx*ny)/xp
# fourier to real space
ψr = abs2(xp*fftshift(ifft(sqrt(nx*ny)ψmom)));
# crop and center around defect
cψr, cx, cy = cropmat(transpose(ψr), sbox.x, sbox.y; xcenter=xdef, ycenter=ydef);
# extract the psi = 1. plane
c = contour(cx, cy, transpose(cψr), 1.0);

# plotting fit
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[1][:imshow](diffctpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])
axes[2][:imshow](diffctidldata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])

for line in c.lines
    xs, ys = coordinates(line)
    for ax in axes
        ax[:plot](xs, ys, "y-")
    end
end

for ax in axes    
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cexpy[1], cexpy[end])
    ax[:set_xlim](cexpx[1], cexpx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
    ax[:set_ylabel](L"$y [\mu m]$")
end

axes[1][:set_title]("pump")
axes[2][:set_title]("idler")
```
