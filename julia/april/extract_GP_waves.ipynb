```julia
fname = basepath*"INPUT";
```

```julia
open(fname,"r") do f
    for line in eachline(f)
        line = strip(replace(line,[',','/'],' '))
        (line[1] == '&' || line[1] == '!') && continue
        include_string("global const " * uppercase(line))
        println(uppercase(line))
    end
end
```

```julia
global const ΔX = 2LX/NX
global const ΔY = 2LY/NY
#conversion factor, in mu
global const lC = 0.868;
```

```julia
sbox = sysbox(-LY:ΔY:LY-ΔY, -LX:ΔX:LX-ΔX, 2pi*DSP.fftshift(DSP.fftfreq(NX, 1/ΔX)), 2pi*DSP.fftshift(DSP.fftfreq(NY, 1/ΔY)))
```

```julia
pump = readdata(basepath, "opo_ph-spc_enfilt_pump.dat", "copy_and_replace.sh");
idler = readdata(basepath, "opo_ph-spc_enfilt_idler.dat", "copy_and_replace.sh");
```

```julia
const xdef = lC * DEF_X_POS
const ydef = lC * DEF_Y_POS;
```

```julia
# read exp data
cpmpdata, cx, cy = cropmat(pump, lC * sbox.x, lC * sbox.y; xcenter=xdef, ycenter=ydef)
cidldata, cx, cy = cropmat(idler, lC * sbox.x, lC * sbox.y; xcenter=xdef, ycenter=ydef);
```

```julia
# Gaussian filtering
sigma = [4, 4]
gausscpmpdata = Images.imfilter_gaussian(cpmpdata, sigma)
diffcpmpdata = cpmpdata - gausscpmpdata
gausscidldata = Images.imfilter_gaussian(cidldata, sigma)
diffcidldata = cidldata - gausscidldata;
```

```julia
# plotting gaussian filtered exp data
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[1,1][:imshow](gausscpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])
axes[1,2][:imshow](gausscidldata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])

axes[2,1][:imshow](diffcpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])
axes[2,2][:imshow](diffcidldata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])

for ax in axes
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cy[1], cy[end])
    ax[:set_xlim](cx[1], cx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
    ax[:set_ylabel](L"$y [\mu m]$")
end

axes[1,1][:set_title]("pump")
axes[2,1][:set_title]("pump")
axes[1,2][:set_title]("idler")
axes[2,2][:set_title]("idler")
```

```julia
y0pmp = 6.
y0idl = 6.
# get corresponding index
idxy0pmp = getidx(y0pmp, maximum(cy), step(cy))
idxy0idl = getidx(y0idl, maximum(cy), step(cy));
```

```julia
# plotting data in 2D
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[1][:imshow](diffcpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])
axes[2][:imshow](diffcidldata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])

axes[1][:plot]((cx[1], cx[end]), (y0pmp, y0pmp), "k-")
axes[2][:plot]((cx[1], cx[end]), (y0idl, y0idl), "k-")

for ax in axes
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cy[1], cy[end])
    ax[:set_xlim](cx[1], cx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
    ax[:set_ylabel](L"$y [\mu m]$")
end

axes[1][:set_title]("pump")
axes[2][:set_title]("idler")
```

```julia
# extract the psi = ct. plane
levelpmp = 1.2
levelidl = -.02;
```

```julia
# plotting cuts through exp data
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[1][:plot](cx, vec(diffcpmpdata[idxy0pmp, :])) 
axes[1][:grid]()

axes[2][:plot](cx, vec(diffcidldata[idxy0idl, :])) 
axes[2][:grid]()

axes[1][:plot]((cx[1], cx[end]), (levelpmp, levelpmp), "k-")
axes[2][:plot]((cx[1], cx[end]), (levelidl, levelidl), "k-")

for ax in axes    
    ax[:set_xlim](cx[1], cx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
end

axes[1][:set_title]("pump")
axes[2][:set_title]("idler")
```

```julia
# get contour level at z==level
cpmp = contour(cx, cy, transpose(diffcpmpdata), levelpmp);
cidl = contour(cx, cy, transpose(diffcidldata), levelidl);
```

```julia
# plotting waves on top
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[1][:imshow](diffcpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])
axes[2][:imshow](diffcidldata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])

plotcontours(cpmp, axes[1])
plotcontours(cpmp, axes[2])

for ax in axes    
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cy[1], cy[end])
    ax[:set_xlim](cx[1], cx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
    ax[:set_ylabel](L"$y [\mu m]$")
end

axes[1][:set_title]("pump")
axes[2][:set_title]("idler")
```
