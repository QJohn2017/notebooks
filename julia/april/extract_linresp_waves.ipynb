```julia
global const ΔX = 2LX/NX
global const ΔY = 2LY/NY;
```

```julia
const xdef = lC * DEF_X_POS
const ydef = lC * DEF_Y_POS;
```

```julia
const sbox = sysbox(-LY:ΔY:LY-ΔY, -LX:ΔX:LX-ΔX, 2pi*DSP.fftshift(DSP.fftfreq(NX, 1/ΔX)), 2pi*DSP.fftshift(DSP.fftfreq(NY, 1/ΔY)));
```

```julia
const pump = readdata(basepath, "opo_ph-spc_enfilt_pump.dat", "copy_and_replace.sh", NY);
const idler = readdata(basepath, "opo_ph-spc_enfilt_idler.dat", "copy_and_replace.sh", NY);
```

```julia
# read numerical data
const cpmpdata, cx, cy = cropmat(pump, lC * sbox.x, lC * sbox.y; xcenter=xdef, ycenter=ydef)
const cidldata, cx, cy = cropmat(idler, lC * sbox.x, lC * sbox.y; xcenter=xdef, ycenter=ydef);
```

```julia
# Gaussian filtering
const gausscpmpdata, diffcpmpdata = filtergauss(cpmpdata, [4., 4.])
const gausscidldata, diffcidldata = filtergauss(cidldata, [4., 4.]);
```

```julia
# plotting cuts through data
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

sliceplot(vec(axes[1,:]), diffcpmpdata, cx, cy, "pump"; y0=6., level=1.45)
const cpmp = contour(cx, cy, transpose(diffcpmpdata), 1.45);

sliceplot(vec(axes[2,:]), diffcidldata, cx, cy, "idler"; y0=6., level=-2e-2)
const cidl = contour(cx, cy, transpose(diffcidldata), -2e-2);

for ax in axes
    ax[:set_xlim](cx[1], cx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
end

for ax in axes[:, 2]
    ax[:grid]()
end

for ax in axes[:, 1]
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cy[1], cy[end])
    ax[:set_ylabel](L"$y [\mu m]$")
end
```

```julia
# plotting waves on top
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[1][:imshow](diffcpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])
axes[1][:set_title]("pump")

axes[2][:imshow](diffcidldata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])
axes[2][:set_title]("idler")

for ax in axes
    plotcontours(cpmp, ax)
end

for ax in axes    
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cy[1], cy[end])
    ax[:set_xlim](cx[1], cx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
    ax[:set_ylabel](L"$y [\mu m]$")
end
```
