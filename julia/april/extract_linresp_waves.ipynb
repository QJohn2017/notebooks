```julia
# modules
using PyPlot
using Contour
using IniFile
```

```julia
# my modules
using Various
```

```julia
ini = Inifile()
read(ini, "/home/berceanu/notebooks/OPODrag/parameters.ini");
```

```julia
# ks0.7 ks0.0 ks-0.4
const kssect = "ks-0.4"; 
```

```julia
const Δk  = get_float(ini, kssect, "delta_k")
const nkx = get_int(ini, kssect, "nkx");
```

```julia
const sider = pi/Δk
const sidek = nkx*Δk/2
const Δr    = pi/sidek;
```

```julia
const xdef = ydef =0.;
```

```julia
const sbox = sysbox(-sider:Δr:sider-Δr, -sider:Δr:sider-Δr, -sidek:Δk:sidek-Δk, -sidek:Δk:sidek-Δk)
```

```julia
resr = readdata("/home/berceanu/notebooks/OPODrag/", "OPODrag.py", "res_r.npy", kssect);
```

```julia
const pump = resr[:,:,2]
const idler = resr[:,:,3];
```

```julia
# crop data
const cpmpdata, cx, cy = cropmat(pump, sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=15.)
const cidldata, cx, cy = cropmat(idler, sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=15.);
```

```julia
# Gaussian filtering
const gausscpmpdata, diffcpmpdata = filtergauss(cpmpdata, [4., 4.])
const gausscidldata, diffcidldata = filtergauss(cidldata, [4., 4.]);
```

```julia
# plotting cuts through data
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

sliceplot(vec(axes[1,:]), diffcpmpdata, cx, cy, "pump"; y0=0., level=0.004)
const cpmp = contour(cx, cy, transpose(diffcpmpdata), 0.);

sliceplot(vec(axes[2,:]), diffcidldata, cx, cy, "idler"; y0=0., level=0.)
const cidl = contour(cx, cy, transpose(diffcidldata), 0.);

for ax in axes
    ax[:set_xlim](cx[1], cx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
end

for ax in axes[:, 2]
    ax[:grid]()
end

for ax in axes[:, 1]
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cy[1], cy[end])
    ax[:set_ylabel](L"$y [\mu m]$")
end
```

```julia
# plotting waves on top
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[1][:imshow](diffcpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])
axes[1][:set_title]("pump")

axes[2][:imshow](diffcidldata, ColorMap("gray"), origin="lower",
                            extent=[cx[1], cx[end], cy[1], cy[end]])
axes[2][:set_title]("idler")

for ax in axes
    plotcontours(cpmp, ax)
end

for ax in axes    
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cy[1], cy[end])
    ax[:set_xlim](cx[1], cx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
    ax[:set_ylabel](L"$y [\mu m]$")
end
```
