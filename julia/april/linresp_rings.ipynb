```julia
ini = Inifile()
read(ini, "/home/berceanu/notebooks/OPODrag/parameters.ini");
```

```julia
# ks0.7 ks0.0 ks-0.4
const kssect = "ks0.0"; 
```

```julia
const Δk  = get_float(ini, kssect, "delta_k")
const nkx = get_int(ini, kssect, "nkx");
```

```julia
const sider = pi/Δk
const sidek = nkx*Δk/2
const Δr    = pi/sidek;
```

```julia
const xdef = ydef =0.;
```

```julia
const sbox = sysbox(-sider:Δr:sider-Δr, -sider:Δr:sider-Δr, -sidek:Δk:sidek-Δk, -sidek:Δk:sidek-Δk)
```

```julia
psi_k = readdata("/home/berceanu/notebooks/OPODrag/", "opodrag.py", "psi_k.npy", kssect);
```

```julia
const signal = psi_k[:,:,1]
const pump = psi_k[:,:,2]
const idler = psi_k[:,:,3];
```

```julia
# crop data
csigdata, ckx, cky = cropmat(signal, sbox.kx, sbox.ky; side=8.)
cpmpdata, ckx, cky = cropmat(pump, sbox.kx, sbox.ky; side=8.)
cidldata, ckx, cky = cropmat(idler, sbox.kx, sbox.ky; side=8.);
```

```julia
# plotting waves on top
fig, ax = plt.subplots(figsize=(4, 4))

ax[:imshow](log10(abs2(csigdata)), ColorMap("gray"), origin="lower",
extent=[ckx[1], ckx[end], cky[1], cky[end]])

ax[:set_ylim](cky[1], cky[end])
ax[:set_xlim](ckx[1], ckx[end])
ax[:set_ylabel](L"$k_y [\mu m^{-1}]$")
ax[:set_xlabel](L"$k_x [\mu m^{-1}]$")
```

```julia
# plotting cuts through data
fig, axes = plt.subplots(3, 2, figsize=(10, 10))

sliceplot(vec(axes[1,:]), clamp(abs2(csigdata),0,1e-05), ckx, cky, "signal"; y0=0., level=0.)

sliceplot(vec(axes[2,:]), clamp(abs2(cpmpdata),0,1e-05), ckx, cky, "pump"; y0=0., level=0.)

sliceplot(vec(axes[3,:]), clamp(abs2(cidldata),0,1e-05), ckx, cky, "idler"; y0=0., level=0.)

for ax in axes
    ax[:set_xlim](ckx[1], ckx[end])
    ax[:set_xlabel](L"$k_x [\mu m^{-1}]$")
end

for ax in axes[:, 2]
    ax[:grid]()
end

for ax in axes[:, 1]
    ax[:set_ylim](cky[1], cky[end])
    ax[:set_ylabel](L"$k_y [\mu m^{-1}]$")
end
```
