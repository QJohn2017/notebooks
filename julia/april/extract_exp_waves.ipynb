```julia
# modules
using PyPlot
using Contour
using Images
```

```julia
# my modules
using Various
```

```julia
# exp params
const NY = 200
const NX = 1024
const Δy = 0.64
const Δx = 0.16

const y = sider(;l=(NY-1)*Δy/2, n=NY)
const x = sider(;l=(NX-1)*Δx/2, n=NX)

const xdef, ydef = -11.2, -0.2

# read exp data
ctpmpdata, cexpx, cexpy = gettrunccrop("E-pump.dat", x, y; xcenter=xdef, ycenter=ydef)
ctidldata, cexpx, cexpy = gettrunccrop("E-idler.dat", x, y; xcenter=xdef, ycenter=ydef);
```

```julia
# Gaussian filtering
sigma = [4, 4]
gaussctpmpdata = Images.imfilter_gaussian(ctpmpdata, sigma)
diffctpmpdata = ctpmpdata - gaussctpmpdata
gaussctidldata = Images.imfilter_gaussian(ctidldata, sigma)
diffctidldata = ctidldata - gaussctidldata;

# plotting gaussian filtered exp data
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[1,1][:imshow](gaussctpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])
axes[1,2][:imshow](gaussctidldata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])

axes[2,1][:imshow](diffctpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])
axes[2,2][:imshow](diffctidldata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])

for ax in axes
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cexpy[1], cexpy[end])
    ax[:set_xlim](cexpx[1], cexpx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
    ax[:set_ylabel](L"$y [\mu m]$")
end

axes[1,1][:set_title]("pump")
axes[2,1][:set_title]("pump")
axes[1,2][:set_title]("idler")
axes[2,2][:set_title]("idler")
```

```julia
y0pmp = 10.
# fill ct. array for plotting
vy0pmp = fill(y0pmp, size(cexpx));
# get corresponding index
idxy0pmp = getidx(y0pmp, maximum(cexpy), step(cexpy))
# check
cexpy[idxy0pmp] - y0pmp #0
```

```julia
y0idl = 15.
# fill ct. array for plotting
vy0idl = fill(y0idl, size(cexpx));
# get corresponding index
idxy0idl = getidx(y0idl, maximum(cexpy), step(cexpy))
# check
cexpy[idxy0idl] - y0idl #0
```

```julia
# plotting exp data in 2D
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[1][:imshow](diffctpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])
axes[2][:imshow](diffctidldata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])

axes[1][:plot](cexpx, vy0pmp)
axes[2][:plot](cexpx, vy0idl)

for ax in axes
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cexpy[1], cexpy[end])
    ax[:set_xlim](cexpx[1], cexpx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
    ax[:set_ylabel](L"$y [\mu m]$")
end

axes[1][:set_title]("pump")
axes[2][:set_title]("idler")
```

```julia
# extract the psi = 1. plane
levelpmp = 20.
levelidl = 5.
vlevelpmp = fill(levelpmp, size(cexpx))
vlevelidl = fill(levelidl, size(cexpx));
```

```julia
# plotting cuts through exp data
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[1][:plot](cexpx, vec(diffctpmpdata[idxy0pmp, :])) 
axes[1][:grid]()

axes[2][:plot](cexpx, vec(diffctidldata[idxy0idl, :])) 
axes[2][:grid]()

axes[1][:plot](cexpx, vlevelpmp)
axes[2][:plot](cexpx, vlevelidl)

for ax in axes    
    #ax[:set_ylim](cexpy[1], cexpy[end])
    ax[:set_xlim](cexpx[1], cexpx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
    #ax[:set_ylabel](L"$y [\mu m]$")
end

axes[1][:set_title]("pump")
axes[2][:set_title]("idler")
```

```julia
cpmp = contour(cexpx, cexpy, transpose(diffctpmpdata), levelpmp);
cidl = contour(cexpx, cexpy, transpose(diffctidldata), levelidl);
```

```julia
# plotting waves on top
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[1][:imshow](diffctpmpdata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])
axes[2][:imshow](diffctidldata, ColorMap("gray"), origin="lower",
                            extent=[cexpx[1], cexpx[end], cexpy[1], cexpy[end]])

for line in cpmp.lines
    xs, ys = coordinates(line)
    axes[1][:plot](xs, ys, "y-")
end

for line in cidl.lines
    xs, ys = coordinates(line)
    axes[2][:plot](xs, ys, "y-")
end

for ax in axes    
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cexpy[1], cexpy[end])
    ax[:set_xlim](cexpx[1], cexpx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
    ax[:set_ylabel](L"$y [\mu m]$")
end

axes[1][:set_title]("pump")
axes[2][:set_title]("idler")
```
