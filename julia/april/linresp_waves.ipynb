```julia
# modules
using PyPlot
using Contour
using IniFile
```

```julia
# my modules
using Various
```

```julia
ini = Inifile()
read(ini, "/home/berceanu/notebooks/OPODrag/parameters.ini");
```

```julia
# ks0.7 ks0.0 ks-0.4
const kssect = "ks0.0"; 
```

```julia
const Δk  = get_float(ini, kssect, "delta_k")
const nkx = get_int(ini, kssect, "nkx");
```

```julia
const sider = pi/Δk
const sidek = nkx*Δk/2
const Δr    = pi/sidek;
```

```julia
const xdef = ydef =0.;
```

```julia
const sbox = sysbox(-sider:Δr:sider-Δr, -sider:Δr:sider-Δr, -sidek:Δk:sidek-Δk, -sidek:Δk:sidek-Δk)
```

```julia
resr = readdata("/home/berceanu/notebooks/OPODrag/", "opodrag.py", "res_r.npy", kssect);
```

```julia
const signal = resr[:,:,1]
const pump = resr[:,:,2]
const idler = resr[:,:,3];
```

```julia
# crop data
csigdata, csigx, csigy = cropmat(signal, sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=20.);
```

```julia
cpmpdata, cx, cy = cropmat(pump, sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=60.);
```

```julia
cidldata, cx, cy = cropmat(idler, sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=60.);
```

```julia
# Gaussian filtering
gausscsigdata, diffcsigdata = filtergauss(csigdata, [4., 4.])
gausscpmpdata, diffcpmpdata = filtergauss(cpmpdata, [4., 4.])
gausscidldata, diffcidldata = filtergauss(cidldata, [4., 4.]);
```

```julia
# plotting waves on top
fig, ax = plt.subplots(figsize=(4, 4))

ax[:imshow](clamp(csigdata,0.8,1.2), ColorMap("gray"), origin="lower",
extent=[csigx[1], csigx[end], csigy[1], csigy[end]])

ax[:set_xticklabels]([])                                                                                                  

ax[:set_ylim](csigy[1], csigy[end])
ax[:set_xlim](csigx[1], csigx[end])
ax[:set_ylabel](L"$y [\mu m]$")
```

```julia
# plotting cuts through data
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

sliceplot(vec(axes[1,:]), diffcpmpdata, cx, cy, "pump"; y0=0., level=0.)
sliceplot(vec(axes[2,:]), diffcidldata, cx, cy, "idler"; y0=0., level=0.)

for ax in axes
    ax[:set_xlim](cx[1], cx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
end

for ax in axes[:, 2]
    ax[:grid]()
end

for ax in axes[:, 1]
    ax[:scatter](xdef, ydef, s=40, c="orange", marker="o")
    ax[:set_ylim](cy[1], cy[end])
    ax[:set_ylabel](L"$y [\mu m]$")
end
```
