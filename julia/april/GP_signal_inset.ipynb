```julia
global const ΔX = 2LX/NX
global const ΔY = 2LY/NY;
```

```julia
#conversion factor, in mu
global const lC = 0.868;
```

```julia
const xdef = lC * DEF_X_POS
const ydef = lC * DEF_Y_POS;
```

```julia
const sbox = sysbox(-LY:ΔY:LY-ΔY, -LX:ΔX:LX-ΔX, 2pi*DSP.fftshift(DSP.fftfreq(NX, 1/ΔX)), 2pi*DSP.fftshift(DSP.fftfreq(NY, 1/ΔY)));
```

```julia
const signal = readdata(basepath, "opo_ph-spc_enfilt_signal.dat", "copy_and_replace.sh", NY);
```

```julia
# read numerical data
csigdata, cx, cy = cropmat(signal, lC * sbox.x, lC * sbox.y; xcenter=xdef, ycenter=ydef, side=7.5);
```

```julia
# Gaussian filtering
gausscsigdata, diffcsigdata = filtergauss(csigdata, [1., 1.]);
```

```julia
# plotting cuts through data
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

sliceplot(vec(axes[:]), diffcsigdata, cx, cy, "signal"; y0=ydef, level=0.)

for ax in axes
    ax[:set_xlim](cx[1], cx[end])
    ax[:set_xlabel](L"$x [\mu m]$")
end

axes[2][:grid]()

axes[1][:scatter](xdef, ydef, s=40, c="orange", marker="o")
axes[1][:set_ylim](cy[1], cy[end])
axes[1][:set_ylabel](L"$y [\mu m]$")
```
