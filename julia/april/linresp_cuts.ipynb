```julia
ini = Inifile()
read(ini, "/home/berceanu/notebooks/OPODrag/parameters.ini");
```

```julia
# ks0.7 ks0.0 ks-0.4
const kssect = "ks0.7"; 
```

```julia
const Δk  = get_float(ini, kssect, "delta_k")
const nkx = get_int(ini, kssect, "nkx");
```

```julia
const sider = pi/Δk
const sidek = nkx*Δk/2
const Δr    = pi/sidek;
```

```julia
const xdef = ydef =0.;
```

```julia
const sbox = sysbox(-sider:Δr:sider-Δr, -sider:Δr:sider-Δr, -sidek:Δk:sidek-Δk, -sidek:Δk:sidek-Δk)
```

```julia
resr = readdata("/home/berceanu/notebooks/OPODrag/", "OPODrag.py", "res_r.npy", kssect);
```

```julia
const signal = resr[:,:,1]
const pump = resr[:,:,2]
const idler = resr[:,:,3];
```

```julia
# crop data
const csigdata, cx, cy = cropmat(signal, sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=60.)
const cpmpdata, cx, cy = cropmat(pump, sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=60.)
const cidldata, cx, cy = cropmat(idler, sbox.x, sbox.y; xcenter=xdef, ycenter=ydef, side=60.);
```

```julia
# Gaussian filtering
const gausscsigdata, diffcsigdata = filtergauss(csigdata, [4., 4.])
const gausscpmpdata, diffcpmpdata = filtergauss(cpmpdata, [4., 4.])
const gausscidldata, diffcidldata = filtergauss(cidldata, [4., 4.]);
```
