```julia
using Reactive, Interact
using PyPlot
```

```julia
x = linspace(0,2π,10);
y = cos(2*x + sin(2*x + 2))
f, axes = plt.subplots(1,2, figsize=(4, 4))
        axes[1][:plot](x, y)
        axes[2][:plot](x, 1./y)
```

```julia
x = linspace(0,2π,1000);
f, axes = plt.subplots(1,2, figsize=(4, 4))
@manipulate for α=1:0.1:3, β=1:0.1:3, γ=1:0.1:3; withfig(f, clear=false) do
        y = cos(α*x + sin(β*x + γ))
        axes[1][:cla]()
        axes[2][:cla]()
        axes[1][:plot](x, y)
        axes[2][:plot](x, 1./y)
    end
end
```

```julia
f = figure();
x = linspace(0,10,10000000)
vertval=slider(0:1:10)
replot = button("Replot") # Commit your changes
map(display, [vertval, replot]) # optional
coeffs = sampleon(replot, lift(tuple, vertval))
PyPlot.plot(x, sin(x))
@manipulate for val=coeffs; withfig(f) do
        axvline(val[1])
    end
end
```

```julia
f = figure()
x = linspace(0,2π,1000)
@manipulate for α=1:0.1:3, β=1:0.1:3, γ=1:0.1:3, leg="a funny plot"; withfig(f) do
        PyPlot.plot(x, cos(α*x + sin(β*x + γ)))
        legend([leg])
    end
end;
```
