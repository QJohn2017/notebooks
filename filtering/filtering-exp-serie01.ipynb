## Tukey window

```python
def tukeywin(window_length, alpha=0.5):
    '''The Tukey window, also known as the tapered cosine window, can be regarded as a cosine lobe of width \alpha * N / 2
    that is convolved with a rectangle window of width (1 - \alpha / 2). At \alpha = 1 it becomes rectangular, and
    at \alpha = 0 it becomes a Hann window. 
    '''
    # Special cases
    if alpha <= 0:
        return np.ones(window_length) #rectangular window
    elif alpha >= 1:
        return np.hanning(window_length)
 
    # Normal case
    x = np.linspace(0, 1, window_length)
    w = np.ones(x.shape)
 
    # first condition 0 <= x < alpha/2
    first_condition = x<alpha/2
    w[first_condition] = 0.5 * (1 + np.cos(2*np.pi/alpha * (x[first_condition] - alpha/2) ))
 
    # second condition already taken care of
 
    # third condition 1 - alpha / 2 <= x <= 1
    third_condition = x>=(1 - alpha/2)
    w[third_condition] = 0.5 * (1 + np.cos(2*np.pi/alpha * (x[third_condition] - 1 + alpha/2))) 
 
    return w
```

```python
tw_x = tukeywin(20,0.3)
tw_y = tukeywin(20,0.3)
tw_2d = np.outer(tw_y,tw_x)
```

# Arbitrary image slice

```python
def sliceImage(I, a, b, *arg, **kws):
    from numpy.linalg import norm
    from scipy.ndimage import map_coordinates
    a = np.array(a)
    b = np.array(b)
    dst = norm(b - a)
    return map_coordinates(I, (a[:,np.newaxis] * np.linspace(1, 0, int(np.ceil(dst)) + 1) +
                               b[:,np.newaxis] * np.linspace(0, 1, int(np.ceil(dst)) + 1)),
                               *arg, **kws)
```

```python
def fft_fit(cut):
    num_samples = cut.size
    x = np.arange(num_samples)
    d_tot = x[-1] #pixels
    sample_rate = num_samples / float(d_tot)  # points/ pixel

    # Taking the fourier transform
    cut_fft = np.fft.rfft(cut)

    # Getting inverse-distances on x-axis right
    rinvdst = [(i*1.0/num_samples)*sample_rate for i in range(num_samples/2+1)]
    
    cut_fft_out = np.abs(cut_fft)/cut_fft.size
    wave_amplitude = np.amax(cut_fft_out)
    k = rinvdst[np.argmax(cut_fft_out)] 
    label = r'$ %.2f \, \sin( %.2f \, 2\pi\,r)$' % (wave_amplitude, k)
    
    fit = wave_amplitude * np.sin(k * x * 2*np.pi)

    return (label, fit, rinvdst, cut_fft_out)
```

```python
url = "https://gist.github.com/andreiberceanu/7141843/raw/0b9d50d3d417b1cbe651560470c098700df5a1fc/image.dat"
f = urllib.urlopen(url)
data = np.loadtxt(f)
```

```python
r0, c0 =  0, 275
r1, c1 = 90, 350
wave = sliceImage(data, (r0,c0), (r1,c1))
lbl, fitted, listofk, wave_fft = fft_fit(wave)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(13.3,10))

axes[0].imshow(data, cmap=cm.gray)
axes[0].plot([c0, c1], [r0, r1], 'ro-')
axes[0].axis('image')

axes[1].plot(wave)
axes[1].set_xlabel('r [pixels]')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(13.3,10))

axes[0].plot(listofk, wave_fft, '-o')

axes[0].set_xlabel('k [1/pixel]')
axes[0].set_ylabel('magnitude')

axes[1].plot(wave, '-')
axes[1].plot(fitted, '--', label=lbl)
axes[1].legend()

axes[1].set_xlabel('r [pixels]')
axes[1].set_ylabel(r'$\left|\psi^i(r)\right|^2$')
```
