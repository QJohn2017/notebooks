I have used the ansatz from http://arxiv.org/abs/cond-mat/0606755v1 and expressed the momentum pertubation in terms of the functions $u$ and $v$.

$$\widetilde{\psi}(k) = e^{-i\omega_p t} \left[\psi^{\text{ss}}_p\delta(k-k_p) + 
\widetilde{u}_p(k-k_p) e^{-i \omega t} + \widetilde{v}^*_p(k_p-k) e^{i \omega 
t}\right]$$

In order to plot the real-space response, I make use of the Fourier shift theorem

$$\psi(r)=\sum_{k}\widetilde{\psi}(k)e^{ikr}=e^{ik_{p}r}\sum_{k}\widetilde{\psi}(k+k_{p})e^{ikr}$$

Therefore, the quantity that I plot in all momentum-space images is

$$\vert\widetilde{\psi}(k+k_{p})\vert^2$$

In view of your previous email, if your

$$\phi_{0}(k)=\phi_{C}^{ss}\delta(k-k_{p})e^{-i\omega_{p}t}$$

then we have the mapping

$$\delta\phi(k)\rightarrow\widetilde{u}_{p}(k-k_{p})+\widetilde{v}_{p}^{*}(k_{p}-k)$$

What I plot is actually the shifted response

$$\delta\phi(k+k_{p})\rightarrow\widetilde{u}_{p}(k)+\widetilde{v}_{p}^{*}(-k)$$
